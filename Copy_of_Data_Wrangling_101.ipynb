## Extract Date Time

```python
df["birth"]
```

```python
df['birth'] = pd.to_datetime(df['birth'], format='%m/%d/%Y')
```

```python
df["birth"]
```

```python
2025 - df["birth"].dt.year
```

```python
df["age"] = 2025 - df["birth"].dt.year
```

```python
df
```

**Quiz !!**

> สร้างคอลัมน์ birth_dow เพื่อแสดงวันเกิดในสัปดาห์ sun, mon, tue, wed, thu, fri, sat

```python
days = ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat']
df['birth_dow'] = df['birth'].dt.day_of_week.apply(lambda x: days[x])
df
```
